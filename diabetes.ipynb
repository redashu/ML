```python
import  pandas  as pd
from sklearn.neighbors  import KNeighborsClassifier
import  seaborn  as snb
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
train=df.loc[:,df.columns != 'Outcome']
#train_test_split(df.loc[:,df.columns != 'Outcome'],)
test=df['Outcome']
train_x,test_x,train_y,test_y=train_test_split(train,test,test_size=0.1)
```

```python
# calling  KNN
algo=KNeighborsClassifier(n_neighbors=5)
trained=algo.fit(train_x,train_y)
trained.predict(test_x)
```
